```python
import seaborn as sns
```

```python
df = pd.read_csv("Datasets/car_sales_data.csv")
```

```python
#quick look at models and prices

df[["Model", "Price"]].head()
df["Model"].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Model", y="Price", data=df, palette= "Set2", hue= "Manufacturer")
plt.title("PRICE DISTRIBUTION BY CAR MODEL", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("PRICE")
plt.xlabel("CAR MODEL")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Model", y="Price", data=df, estimator=pd.Series.mean, errorbar=None, palette= "Set2", hue= "Manufacturer")
plt.title("AVERAGE PRICE BY CAR MODEL", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("AVERAGE PRICE")
plt.xlabel("MODEL")
plt.show()
```
